```python
churnData = pd.read_csv("C:/Users/user/OneDrive/khaishenchan@gmail.com/OneDrive/FYP-Apr 2024/FYP_Project/e-commerce-book-shop-back-end/ChurnTrainingDataset/ecom-user-churn-data.csv")
churnData.head()
```

```python
minMaxData = {
    "column": [],
    "minimum": [],
    "maximum": []
}

for column in churnData.columns:
    minimum = churnData[column].min()
    maximum = churnData[column].max()
    
    minMaxData["column"].append(column)
    minMaxData["minimum"].append(minimum)
    minMaxData["maximum"].append(maximum)

minMaxDataframe = pd.DataFrame(minMaxData)
print(minMaxDataframe)
```
