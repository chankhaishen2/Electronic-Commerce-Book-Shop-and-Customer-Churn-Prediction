```python
minMaxData = {
    "column": [],
    "minimum": [],
    "maximum": []
}

for column in churnData.columns:
    minimum = churnData[column].min()
    maximum = churnData[column].max()
    
    minMaxData["column"].append(column)
    minMaxData["minimum"].append(minimum)
    minMaxData["maximum"].append(maximum)

minMaxDataframe = pd.DataFrame(minMaxData)
print(minMaxDataframe)
```

```python
churnData_selected = pd.DataFrame(churnData, columns = ['ses_rec', 'ses_rec_avg', 'ses_rec_sd', 
                                                        'ses_rec_cv', 'user_rec', 'ses_n', 'int_n',
                                                        'int_n_r', 'tran_n', 'tran_n_r', 'rev_sum',
                                                        'rev_sum_r', 'target_class'])
churnData_selected.head()
```

```python
churnDataArray = np.array(churnData)
print(churnDataArray)
```

```python
x = churnDataArray[:, 0:12]
y = churnDataArray[:, -1]
print(x, y)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print(x_train, x_test, y_train, y_test)
```

```python
from keras import Sequential
from keras.layers import Dense

model = Sequential([
    Dense(50, activation="relu", kernel_regularizer='l2'),
    Dense(30, activation="relu", kernel_regularizer='l2'),
    Dense(15, activation="relu", kernel_regularizer='l2'),
    Dense(1, activation="sigmoid", kernel_regularizer='l2')
])

model.compile(
    loss='binary_crossentropy',
    metrics=['accuracy']
)

history = model.fit(x_train, y_train, batch_size=100, epochs=100, validation_split=0.2)
```

```python
print('loss:', history.history['loss'][-1])
print('accuracy', history.history['accuracy'][-1])
```

```python
results = model.evaluate(x_test, y_test)
print('loss:', results[0])
print('accuracy:', results[1])
```
